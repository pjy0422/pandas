**Advanced Filtering with between(), isin()**

```python
summer = pd.read_csv('summer.csv')
summer.head()
```

```python
og_1988 = summer.loc[summer.Year == 1988]
```

```python
og_1988.head()
```

```python
og_1988.info()
```

```python
og_since1992 = summer.loc[summer.Year>=1992]
og_since1992.head()
```
